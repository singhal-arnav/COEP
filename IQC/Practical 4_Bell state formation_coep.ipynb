```python
#|psi+>

# Create a Quantum Circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 1: Apply X gate on qubit 1 to flip it to |1>
qc.x(1)

# Step 2: Apply Hadamard gate on qubit 0 to create superposition
qc.h(0)

# Step 3: Apply CNOT gate with control qubit 0 and target qubit 1 to create entanglement
qc.cx(0, 1)

# Step 4: Measure both qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
qc.draw('mpl')
```

```python
#|psi->

# Create a Quantum Circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 1: Apply X gate on qubit 1 to flip it to |1>
qc.x(1)

# Step 2: Apply Hadamard gate on qubit 0 to create superposition
qc.h(0)

# Step 3: Apply CNOT gate with control qubit 0 and target qubit 1 to create entanglement
qc.cx(0, 1)

# Step 4: Apply Z gate on qubit 0 to introduce the phase needed for |psi->
qc.z(0)

# Step 5: Measure both qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
qc.draw('mpl')
```

```python
#|phi+>

# Create a Quantum Circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 1: Apply Hadamard gate on qubit 0 to create superposition
qc.h(0)

# Step 2: Apply CNOT gate with control qubit 0 and target qubit 1 to create entanglement
qc.cx(0, 1)

# Step 3: Measure both qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
qc.draw('mpl')
```

```python
#|phi->

qc_phi_minus = QuantumCircuit(2, 2)

# Step 1: Apply Hadamard on qubit 0
qc_phi_minus.h(0)

# Step 2: Apply CNOT with control qubit 0, target qubit 1
qc_phi_minus.cx(0, 1)

# Step 3: Apply Z gate on qubit 0 to introduce the phase (minus sign)
qc_phi_minus.z(0)

# Step 4: Measure both qubits
qc_phi_minus.measure([0, 1], [0, 1])

# Draw the circuit
qc_phi_minus.draw('mpl')
```
