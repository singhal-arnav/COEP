```python
from qiskit_ibm_runtime import QiskitRuntimeService

service = QiskitRuntimeService.save_account(
    channel="ibm_cloud",    # or "ibm_quantum"
    token=token,            # IBM Cloud API key
    instance="open-instance",       # <-- this should be your *CRN* or instance name
    name="default",      # optional
    set_as_default=True,
    overwrite=True
)
```

```python
service = QiskitRuntimeService()
print(service.backends())
```

```python
service.backends(simulator=False, operational=True, min_num_qubits=5)
for b in service.backends():
    status = b.status()
    conf = b.configuration()
    print(f"Backend: {b.name}")
    print(f"  - Operational: {status.operational}")
    print(f"  - Qubits: {conf.n_qubits}")
    print(f"  - Status Message: {status.status_msg}")
    print("-" * 20)
```

```python
service.least_busy(operational=True, min_num_qubits=5)
```

```python
backend = service.backend("ibm_torino")
 
print(
    f"Name: {backend.name}\n"
    f"Version: {backend.version}\n"
    f"No. of qubits: {backend.num_qubits}\n"
)
```
