```python
import math
a = 2
b = 18
c = 7
x1 = (-b + math.sqrt(b*b - 4*a*c))/(2*a)
x2 = (-b - math.sqrt(b*b - 4*a*c))/(2*a)
print('The roots of the quadratic equation are', x1, 'and', x2)
```

```python
n = int(input('Enter a four digit number:'))

rev = 0

rev+=n%10
n//=10
rev*=10
rev+=n%10
n//=10
rev*=10
rev+=n%10
n//=10
rev*=10
rev+=n%10

print('Reversed number =', rev)
```
