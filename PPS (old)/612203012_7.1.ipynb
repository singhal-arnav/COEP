```python
def binary_search(l, low, high, n):   
   if low <= high:  
      mid = (low + high) // 2  
      if l[mid] == n:   
         return mid   
      elif l[mid] > n:   
         return binary_search(l, low, mid - 1, n)   
      else:   
         return binary_search(l, mid + 1, high, n)   
   else:   
      return -1  
  
lst = [12, 24, 32, 39, 45, 50, 54]  
n = 32  
  
res = binary_search(lst, 0, len(lst)-1, n)   
  
if res != -1:   
   print("Element is present at index", res)  
else:   
   print("Element is not present in list")
```
