```python
print('Enter 1 to find the maximum and second maximum element from a list')
print('Enter 2 to generate prime numbers from a given range in a list')
print('Enter 3 to compute X power Y')
print('Enter 4 to find Kth smallest and biggest element in a list')
choice = int(input('Enter your choice: '))

if choice == 1:
  n = int(input('Enter the number of elements in the list:'))
  l = []
  print('Enter the elements in the list:')
  for i in range(n):
    e = int(input())
    l += [e]
  max = l[0]
  for num in l[1:]:
    if num > max:
      max = num
  print('Maximum element =', max)
  sec_max = l[0]
  for num in l[1:]:
    if num > sec_max and num != max:
      sec_max = num
  print('Second maximum element =', sec_max)

elif choice == 2:
  start = int(input('Enter the lower limit of the range: '))
  end = int(input('Enter the upper limit of the range: '))
  primes = []
  for num in range(start, end+1):
    factors = 0
    for i in range(num):
      if num % i == 0:
        factors += 1
    if factors == 2:
      primes.append(num)
  print('Prime numbers in given range =', primes)

elif choice == 3:
  x = int(input('Enter the value of x: '))
  y = int(input('Enter the value of y: '))
  print('x^y =', x**y)

elif choice == 4:
  n = int(input('Enter the number of elements in the list:'))
  l = []
  print('Enter the elements in the list:')
  for i in range(n):
    e = int(input())
    l += [e]
  l.sort()
  k = int(input('Enter the value of k: '))
  print(k, 'th smallest element = ', l[k-1], sep = '')
  print(k, 'th biggest element = ', l[-k], sep = '')

else:
  print('Invalid input')
```
