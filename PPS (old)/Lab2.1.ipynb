```python
sales = int(input('Enter the total sales of the medical representative:'))
if sales >= 100000:
  basics = 4000
  hra = 20/100 * basics
  da = 110/100 * basics
  conveyance = 500
  incentive = 10/100 * sales
  bonus = 1000
else:
  basics = 4000
  hra = 10/100 * basics
  da = 110/100 * basics
  conveyance = 500
  incentive = 4/100 * sales
  bonus = 500
salary = basics + hra + da + conveyance + incentive + bonus
print('Salary of the medical representative =', salary)
```
