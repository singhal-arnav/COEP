```python
n = int(input('Enter the number of elements: '))
l = []
print('Enter the elements in the list:')
for i in range(n):
  e = int(input())
  l.append(e)

k = int(input('Enter the value to be searched for:'))
flag = 0
for i in range(len(l)):
  if l[i] == k:
    flag = 1
    index = i

if flag == 0:
  print('Key not found in list')
else:
  print('Key found in list at index', index)
```
