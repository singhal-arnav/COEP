```python
#Q4

import csv
import pandas as pd

lists = [['Index', 'Name', 'Age'],
    [2, 'Vishal', 22],
    [1, 'Kushal', 25],
    [1, 'Aman', 24]]
myfile = open('data1.csv', 'w')
w = csv.writer(myfile)
w.writerows(lists)
myfile.close()

df = pd.read_csv('data1.csv')
print(df)

print()
df_custom = df.set_index('Index')
print(df_custom)
```

```python
#Q5

df_custom = df_custom.sort_index()
print(df_custom)
```

```python
#Q6

df_custom = df_custom.sort_values(by=['Index','Age'])
print(df_custom)
```

```python
#Q7

df_custom = df_custom.reset_index()
print(df_custom)
print()
print(df_custom['Name'])
```

```python
#Q10

import pandas as pd
import math

d = {'name': ['Vinay', 'Kushal', 'Aman', 'Rahul', 'Ramesh'],
     'occ' : ['engineer', 'doctor', 'engineer', 'doctor', 'doctor'],
     'salary' : [60000, 70000, 50000, 60000, 65000],
     'experience' : [10, math.nan, 7, math.nan, 9]}
df = pd.DataFrame(d)
df = df.fillna(0)
print(df)
print()
print(df.groupby('occ').agg({'salary':'mean'}))
```
